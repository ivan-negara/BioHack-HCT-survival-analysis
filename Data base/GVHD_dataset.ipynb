```python
import category_encoders as ce

encoder= ce.OrdinalEncoder(cols=['Allogeneic transplant type',
                                 'Allogenic transplant source',
                                 'Conditioning type'],return_df=True,
                           mapping=[{'col':'Allogeneic transplant type', 'mapping':{'Haplo':1,'MRD':2,'MMUD':3,'MUD':4}},
                                    {'col':'Allogenic transplant source', 'mapping':{'BSC':0,'BM':1}},
                                    {'col':'Conditioning type', 'mapping':{'RIC':0,'MAC':1}}])


TR[['Allogeneic transplant type',
    'Allogenic transplant source',
    'Conditioning type']] = encoder.fit_transform(TR[['Allogeneic transplant type',
                                                      'Allogenic transplant source',
                                                      'Conditioning type']])

TR.head()
```
